```python
from src.test_optimizer import run_test
from src.optimizer import GradientDescent, GradientDescentMomentum, AdaGrad, Adam

gradient_descent = GradientDescent()
gradient_descent_momemtum = GradientDescentMomentum()
ada_grad = AdaGrad()
adam = Adam()

print("Run gradient descent")
time_gd, loss_gd, acc_gd, acc_test_gd = run_test(optimizer=gradient_descent)
print("Run gradient descent momentum")
time_gdm, loss_gdm, acc_gdm, acc_test_gdm = run_test(optimizer=gradient_descent_momemtum)
print("Run ada grad")
time_ad, loss_ad, acc_ad, acc_test_ad = run_test(optimizer=ada_grad)
print("Run adam grad")
time_adam, loss_adam, acc_adam, acc_test_adam = run_test(optimizer=adam)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from src.utils import smooth


x = np.arange(200)

plt.figure(0)
plt.title('Time')
plt.plot(x, smooth(time_gd), label='gradient descent')
plt.plot(x, smooth(time_gdm), label='gradient descent momentum')
plt.plot(x, smooth(time_ad), label='ada grad')
plt.plot(x, smooth(time_adam), label='adam')
plt.legend()

plt.figure(1)
plt.title('Accuracy')
plt.plot(x, smooth(acc_gd), label='gradient descent')
plt.plot(x, smooth(acc_gdm), label='gradient descent momentum')
plt.plot(x, smooth(acc_ad), label='ada grad')
plt.plot(x, smooth(acc_adam), label='adam')
plt.legend()

plt.figure(2)
plt.title('Test Accuracy')
plt.plot(x, smooth(acc_test_gd), label='gradient descent')
plt.plot(x, smooth(acc_test_gdm), label='gradient descent momentum')
plt.plot(x, smooth(acc_test_ad), label='ada grad')
plt.plot(x, smooth(acc_test_adam), label='adam')
plt.legend()

plt.figure(3)
plt.title('Loss')
plt.plot(x, smooth(loss_gd), label='gradient descent')
plt.plot(x, smooth(loss_gdm), label='gradient descent momentum')
plt.plot(x, smooth(loss_ad), label='ada grad')
plt.plot(x, smooth(loss_adam), label='adam')
plt.legend()

plt.show()
```
